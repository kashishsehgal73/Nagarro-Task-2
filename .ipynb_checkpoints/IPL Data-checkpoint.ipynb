```python
import numpy as np 
import pandas as pd
import os
from pandas import DataFrame as df
```

### Importing the datasets

```python
deliveries = pd.read_csv('deliveries.csv')
matches = pd.read_csv('matches.csv')
```

### Rising Pune Supergiant is present as 2 teams Rising Pune Supergiant and Rising Pune Supergiants, so replacing the value

```python
matches.replace( 'Rising Pune Supergiant', 'Rising Pune Supergiants',inplace = True)
matches.head()
```

## 1.Which team has won/lost maximum number of matches while chasing?

```python
team_won_while_chasing = matches[(matches['team2'] == matches['winner'])]['team2']
team_won_while_chasing.groupby(team_won_while_chasing).count().sort_values(ascending = False).head()
```

```python
team_lost_while_chasing = matches[(matches['team2'] != matches['winner'])]['team2']
team_lost_while_chasing.groupby(team_lost_while_chasing).count().sort_values(ascending = False).head()
```

## 2.Which toss decision is best for each ground?

```python
chasing_team_won = df(matches.loc[(matches['team2'] == matches['winner'])].groupby('venue').count()['id'])
batting_team_won = df(matches.loc[(matches['team1'] == matches['winner'])].groupby('venue').count()['id'])

#Merging two dataframes
toss_decision = pd.merge(batting_team_won,chasing_team_won, on = 'venue')
toss_decision = toss_decision.rename(columns={'id_x':'batting team won', 'id_y':'chasing team won'})

toss_decision['decision'] = toss_decision.apply(lambda x : 'bat' if x['batting team won'] >= x['chasing team won'] else "field", axis=1)
```

## 5.Which batsmen came to crease but never got a chance to face a ball?

```python
dnb_batsman = df(columns = ['match_no', 'batsman'])
for i,g in deliveries.groupby('match_id'):
    dnb =  list(set(g['non_striker']) - set(g['batsman']))
    if len(dnb): 
        dnb_batsman.loc[len(dnb_batsman)] = [i,str(' , '.join(dnb))]
    else:
        dnb_batsman.loc[len(dnb_batsman)] = [i,'None']
dnb_batsman
```

## 6.Which bowler has hit for maximum number of 1s/2s/3s/4s/6s

```python
bowler_data = deliveries[['bowler', 'total_runs']]
bowler_sixes = bowler_data.loc[(bowler_data['total_runs'] == 6) | (bowler_data['total_runs'] == 7)]
bowler_fours = bowler_data.loc[(bowler_data['total_runs'] == 4)]
bowler_threes = bowler_data.loc[(bowler_data['total_runs'] == 3)]
bowler_twos = bowler_data.loc[(bowler_data['total_runs'] == 2)]
bowler_ones = bowler_data.loc[(bowler_data['total_runs'] == 1)]
bowler_data.merge(bowler_sixes,on='bowler')

print("Sixes :",bowler_sixes['bowler'].value_counts().head(3), sep = '\n', end = '\n\n')
print("Fours :",bowler_fours['bowler'].value_counts().head(3), sep = '\n', end = '\n\n')
print("Threes :",bowler_threes['bowler'].value_counts().head(3), sep = '\n', end = '\n\n')
print("Twos :",bowler_twos['bowler'].value_counts().head(3), sep = '\n', end = '\n\n')
print("Ones :",bowler_ones['bowler'].value_counts().head(3), sep = '\n', end = '\n\n')

```

## 10.Success rate of teams


```python
matches_won = matches.groupby('winner').count()
total_matches = matches['team1'].value_counts()+ matches['team2'].value_counts()

matches_won['Total matches']=total_matches
match_succes_rate = (matches_won["id"]/total_matches)*100

data = match_succes_rate.sort_values(ascending = False)
print(data)
```

## 11.How winning the toss effects the match result

```python
tosswin_win = matches['id'][matches['toss_winner'] == matches['winner']].count()
total_matches=matches['id'].count()
Success_rate = ((matches[matches['toss_winner'] == matches['winner']].count())/(total_matches))*100

print("Number of matches in which Toss winner is the game winner is :",tosswin_win, "out of",total_matches," ie.,", Success_rate["id"],"%" )
```
